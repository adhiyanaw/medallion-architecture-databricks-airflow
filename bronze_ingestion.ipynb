```python
# coin_markets

from pyspark.sql.functions import current_timestamp

base_path = "/Volumes/medallionarchitecture/bronze/crypto_raw/raw"

df_markets = spark.read.json(f"{base_path}/coin_markets/*/data.json")

df_markets = df_markets.withColumn("ingestion_time", current_timestamp())

df_markets.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("medallionarchitecture.bronze.coin_markets")
```

```python
# simple_price

df_price_raw = spark.read.json(f"{base_path}/simple_price/*/data.json")

df_price = df_price_raw.selectExpr(
    "stack(2, 'bitcoin', cast(bitcoin.usd as double), 'ethereum', cast(ethereum.usd as double)) as (coin, price_usd)"
)

from pyspark.sql.functions import current_timestamp
df_price = df_price.withColumn("ingestion_time", current_timestamp())

df_price.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("medallionarchitecture.bronze.simple_price")
```

```python
# coin_details

df_details = spark.read.json(f"{base_path}/coin_details/*/data.json")

from pyspark.sql.functions import current_timestamp
df_details = df_details.withColumn("ingestion_time", current_timestamp())

df_details.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("medallionarchitecture.bronze.coin_details")
```
